```python
decision = {}
for item in raw:
    segs = re.split("-+", item)
    decision[os.path.splitext(segs[0])[0]] = int(segs[1].strip("\n").startswith("A"))
```

## Getting the win-ratio of each advocate for each charge

```python
def avg_win_ratio(case_list, scores):
    return sum([scores[case] for case in case_list]) * 1./len(case_list) if len(case_list) != 0 else 0
```

```python
charge_adv_win_ratios = defaultdict(lambda: dict())
for charge in charges:
    for adv in advs.values():
        case_list = set(adv_cases[adv]).intersection(set(charge_cases[charge]))
        print(case_list)
        charge_adv_win_ratios[charge][adv] = avg_win_ratio(case_list, decision)
        # Order by score
    charge_adv_win_ratios[charge] = {
                                    k: v
                                    for k, v in sorted(
                                        charge_adv_win_ratios[charge].items(),
                                        key=lambda x: x[1], reverse=True)}
```
