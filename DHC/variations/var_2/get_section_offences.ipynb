```python
def separate_elements(df, num):
    raw = df.loc[[num]]["Classification of offences"]
    raw = str(raw)
    elements = raw.split(')')
    elements[0] = re.sub(f"{num}", '', elements[0])
    elements = elements[:-1]
    elements[0] = elements[0].split('.')[-1]
    info = [v.split('(') for v in elements]
    info = [(v[1].strip(), v[0].strip()) for v in info]
    sub_df = pd.DataFrame(data=info, columns=['Sections covered', 'Classification of offences'])
    return sub_df
```

```python
dict_path = "/home/workboots/Datasets/IndiaCode/CentralActs/section_jsons/186045.json"
```

```python
with open(dict_path, 'r') as f:
    ipc = json.load(f)
```

```python
section_names = list(ipc.keys())
```

```python
section_names
```

```python
section_offences = defaultdict(lambda : list())
```

```python
for sec in section_names:
    for off in offences_dict:

        if natsorted([sec, off])[ -1] == sec and natsorted([sec, offences_dict[off]['End']])[0] == sec:
            section_offences[sec].append(offences_dict[off]['Classification of offences'])
```

```python
section_offences
```

```python
section_offences = {
    f"Indian Penal Code, 1860_{k}": v for k, v in section_offences.items()
}
```
